```python
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread('/picture/picture6.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img = cv2.resize(img, (0, 0), fx=1, fy=1)

# 偵測臉
person = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
gray = cv2.medianBlur(gray, 3)
people = person.detectMultiScale(gray, 1.1, 2)
print(len(people))

for (x, y, w, h) in people:
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
# 偵測行人
person = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_fullbody.xml")
gray = cv2.medianBlur(gray, 3)
people = person.detectMultiScale(gray, 1.02, 3)
print(len(people))

for (x, y, w, h) in people:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 0, 255), 2)

cv2_imshow(gray)
cv2_imshow(img)
cv2.waitKey(0)
```
